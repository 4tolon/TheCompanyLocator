###  Busquedas en foursquare , (gracias a Jean Charles Yamada)

```python
def foursquare(busqueda, lat, lon):
    '''
    Funcion para llamar a la api de four square mediante un string de busqueda y una parejade corrdonadas
    '''
    tok_id = os.getenv('fs_cli')
    tok_se = os.getenv('fs_sec')
    url = 'https://api.foursquare.com/v2/venues/search'

    params = dict(
        client_id=tok_id,
        client_secret=tok_se,
        v='20180323',
        ll=f'{lat},{lon}',
        query=f'{busqueda}',
        limit=10,
        radius=1000
    )
    resp = requests.get(url=url, params=params).json()["response"]["venues"]
    return resp
```
